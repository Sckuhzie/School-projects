## Extraction des règles et les mesures d'évaluation :

### leverage :  
- $levarage(A \rightarrow C)=support(A→C)−support(A)×support(C)$ dans l'intervalle  [−1,1]
- Leverage calcule la  différence entre les frequences observées de A et C qui apparaissent ensemble 
vs la fréquence attendue si A et C étaient independants. 
- <font color="red"> Un  leverage = 0 indique l'indépendance des deux.</font>

### lift (intérêt):   
- $lift(A→C)= \frac{confiance(A→C)}{support(C)}$ dans l'intervalle  [0,∞]
- Le Lift est utilisé pour mesurer combien l'antécédent (A) et la conséquence (C) d'une règle A->C apparaissent
ensemble vs. s'ils étaient statistiquement indépendantes. 
- <font color="red"> Si   A et C sont independents, le Lift sera= 1.  </font>
- <font color="red">  Si Lift>1 : association (corrélation) positive (resp. négative). </font>

### conviction :
- $conviction(A→C)= \frac{1−support(C)}{1−confiance(A→C)}$ dans l'intervalle [0,∞]
- <font color="red">  Une conviction élevée signifie que la conséquence de la règle dépend de l'antécédent.</font> 
- P. Ex. , en cas de score maximal de confiance, le dénominateur devient 0 (1-1) pour lequel la conviction ='inf'.
- <font color="red"> De manière similaire à  Lift, si les items sont indépendants, la conviction = 1..</font>
