### Exercice 3

On remarque que pour l'algorithme SimpleKMeans, la valeur moyenne des deux clusters pour l'attribut Play est "yes", alors que l'on a un cluster pour "yes", et l'un pour "no" avec la méthode FarthestFirst. Avec les deux algorithmes, on a 10 éléments dans le premier cluster et 4 dans le deuxième. Dans la base de données Weather, on a 9 éléments dont l'attribut play est "yes" et 5 pour "no". Donc FarthestFirst est mieux pour générer deux clusters, chacun représentant une valeur de Play différente.

## 3. Evaluation par rapport à un attribut de classe

### Exercice 9

| Valeur / Méthode  | Farthest First | SimpleKMeans |
|-------------------|----------------|--------------|
| **Faux positifs** | 1              | 1            |
| **Faux négatifs** | 5              | 5            |
| **Taux d'erreur** | 42,9 %         | 42,9 %       |

### Exercice 10

| Algorithme    | Seed | Taux d'erreur | Faux positifs | Faux négatifs |
|---------------|------|---------------|---------------|---------------|
| SimpleKMeans  | 1    | 35,7 %        | 3             | 2             |
| FarthestFirst | 100  | 35,7 %        | 2             | 3             |

Les deux algorithmes ont le même taux d'erreur. Cependant, comme la base de données de weather possède plus de positifs que de négatifs, on prèfère avoir moins de faux négatifs pour en rater le moins possible. On va donc privilégier l'algorithme SimpleKMeans.

## 4. Choisir un modèle obtenu par K-means

### Exercice 12

|           | Taux d'erreur | Classe cluster 0 | Instances cluster 0 | Classe cluster 1 | Instances cluster 1 |
|-----------|---------------|------------------|---------------------|------------------|---------------------|
| Méthode 1 | 1510          | republican       | 49%                 | democrat         | 51%                 |
| Méthode 2 | 970           | democrat         | 49%                 | republican       | 51%                 |
| Méthode 3 | 1282          | democrat         | 47%                 | republican       | 53%                 |
| Méthode 4 | 1322          | republican       | 45%                 | democrat         | 55%                 |

Ici, toutes les méthodes donnent un cluster pour chaque classe, mais la méthode 2 est celle qui donne l'erreur la plus petites. C'est cette dernière qu'on va donc choisir.
