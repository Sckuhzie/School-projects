# Classification Naive Bayes (avec Scikit-learn)

#### Le Dataset : Météo mais très simplifié (avec 3 colonnes)
#### Exercice : faire de même mais avec la BD météo complet.

```python
#  features et variable cible (décision) 
wheather=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny',
'Rainy','Sunny','Overcast','Overcast','Rainy']
temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']

play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
```

#### Encodage des valeurs 

On encode les valeurs : conventir les str en un entier (qui sera le code).  
Par exemple: 'Overcast', 'Rainy', 'Sunny' deviennent 0, 1, 2.  Valeurs entre 0 ...nb_feautures-1.  
On le fait à l'aide de Scikit-learn et LabelEncoder.

```python
### Pour tout afficher (sans print)
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# Importer LabelEncoder 
from sklearn import preprocessing
#creation  
le = preprocessing.LabelEncoder()
# conversion 
wheather_encoded=le.fit_transform(wheather)
wheather_encoded
```

De même pour les attributs "temp" et "play"

```python
# Convertion  
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)
"Temp:",temp_encoded
"Play:",label
```

l'ensemble donne (liste de tuples) 

```python
# 
features=list(zip(wheather_encoded,temp_encoded))
features
```

#### Le Model  : naive bayes classifier

Etapes

   * Instanciation
   * Fit le dataset  (entrainer)
   * Prédiction

```python
#LE modèle Gaussian Naive Bayes  
from sklearn.naive_bayes import GaussianNB

#Instanciation
model = GaussianNB()

# Apprentissage  
model.fit(features,label)

#Prédiction sur un cas simple : 0:Overcast, 2:Mild
predicted= model.predict([[0,2]])  
"Predicted Value:", predicted
```
